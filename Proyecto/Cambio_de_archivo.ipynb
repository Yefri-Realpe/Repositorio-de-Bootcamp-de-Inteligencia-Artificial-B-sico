# Task
Transform the data in the "rendimiento_diario_conductores.csv" file by removing the date and time columns and converting the values in the "zona" column to numerical representations (e.g., 0 for 'norte', 1 for 'sur', etc.).

## Cargar los datos

### Subtask:
Carga el archivo CSV `rendimiento_diario_conductores.csv` en un DataFrame de pandas.


**Reasoning**:
Load the data into a pandas DataFrame and display the first few rows.



```python
import pandas as pd

df = pd.read_csv('rendimiento_diario_conductores.csv')
display(df.head())
```

## Eliminar columnas

### Subtask:
Elimina las columnas de fecha y hora del DataFrame.


**Reasoning**:
Drop the 'fecha' and 'hora' columns from the DataFrame and display the head to confirm the changes.



```python
df.drop(['fecha', 'hora'], axis=1, inplace=True)
display(df.head())
```

## Transformar la columna 'zona'

### Subtask:
Convierte los valores de la columna 'zona' a números (ej: 0 para norte, 1 para sur, etc.).


**Reasoning**:
Import LabelEncoder, instantiate it, fit it to the 'zona' column, transform the 'zona' column, and display the head of the DataFrame to verify the changes.



```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['zona'] = label_encoder.fit_transform(df['zona'])

display(df)
```

## Guardar los datos

### Subtask:
Guarda el DataFrame modificado en un nuevo archivo CSV.


**Reasoning**:
Save the modified DataFrame to a new CSV file without the index.



```python
df.to_csv('rendimiento_diario_conductores_transformado.csv', index=False)
```

*   The categorical values in the 'zona' column were successfully converted into numerical labels using `LabelEncoder`.

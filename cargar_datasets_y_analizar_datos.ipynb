```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_titanic = pd.read_csv("Titanic-Dataset.csv")

# 1. InformaciÃ³n general
print("InformaciÃ³n del DataFrame:")
df_titanic.info()

# 2. EstadÃ­sticas descriptivas
print("\nEstadÃ­sticas descriptivas:")
df_titanic.describe()

# 3. VisualizaciÃ³n de la distribuciÃ³n de variables categÃ³ricas
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.countplot(x='Survived', data=df_titanic)
plt.title('DistribuciÃ³n de Supervivientes')

plt.subplot(1, 3, 2)
sns.countplot(x='Pclass', data=df_titanic)
plt.title('DistribuciÃ³n por Clase')

plt.subplot(1, 3, 3)
sns.countplot(x='Sex', data=df_titanic)
plt.title('DistribuciÃ³n por Sexo')
plt.tight_layout()
plt.show()

# 4. VisualizaciÃ³n de la distribuciÃ³n de variables numÃ©ricas
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df_titanic['Age'].dropna(), kde=True)
plt.title('DistribuciÃ³n de Edad')

plt.subplot(1, 2, 2)
sns.histplot(df_titanic['Fare'], kde=True)
plt.title('DistribuciÃ³n de Tarifa')
plt.tight_layout()
plt.show()

# 5. AnÃ¡lisis de la relaciÃ³n entre Supervivencia y Edad
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Age', data=df_titanic)
plt.title('Edad por Supervivencia')
plt.show()

# 6. Matriz de correlaciÃ³n (para variables numÃ©ricas)
numeric_cols = df_titanic.select_dtypes(include=['number']).columns
correlation_matrix = df_titanic[numeric_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de CorrelaciÃ³n')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_titanic = pd.read_csv("Titanic-Dataset.csv")

# 1. Información general
print("Información del DataFrame:")
df_titanic.info()

# 2. Estadísticas descriptivas
print("\nEstadísticas descriptivas:")
df_titanic.describe()

# 3. Visualización de la distribución de variables categóricas
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.countplot(x='Survived', data=df_titanic)
plt.title('Distribución de Supervivientes')

plt.subplot(1, 3, 2)
sns.countplot(x='Pclass', data=df_titanic)
plt.title('Distribución por Clase')

plt.subplot(1, 3, 3)
sns.countplot(x='Sex', data=df_titanic)
plt.title('Distribución por Sexo')
plt.tight_layout()
plt.show()

# 4. Visualización de la distribución de variables numéricas
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df_titanic['Age'].dropna(), kde=True)
plt.title('Distribución de Edad')

plt.subplot(1, 2, 2)
sns.histplot(df_titanic['Fare'], kde=True)
plt.title('Distribución de Tarifa')
plt.tight_layout()
plt.show()

# 5. Análisis de la relación entre Supervivencia y Edad
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Age', data=df_titanic)
plt.title('Edad por Supervivencia')
plt.show()

# 6. Matriz de correlación (para variables numéricas)
numeric_cols = df_titanic.select_dtypes(include=['number']).columns
correlation_matrix = df_titanic[numeric_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación')
plt.show()
```

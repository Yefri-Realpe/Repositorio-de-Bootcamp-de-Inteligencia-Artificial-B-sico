```python
import pandas as pd

value_n_list = ["Andres", "Phillip", "John", "Felipe"]
print(value_n_list)

value_w_list = [50, 78, 91.5, 86]
print(value_w_list)

value_h_list =[1.76, 1.87, 1.69, 1.65]
print(value_h_list)
data_dict = {"name":value_n_list,
             "weight":value_w_list,
             "height":value_h_list}
print(data_dict)
df=pd.DataFrame(data_dict)
print(df)
df
```

```python
df["bmi"] = df["weight"] / (df["height"] ** 2)
df
```

```python
df.describe()
```
